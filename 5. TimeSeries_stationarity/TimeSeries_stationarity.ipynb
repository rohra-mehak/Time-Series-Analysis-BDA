```python
import yfinance as yahooFinance
import datetime
import pandas as pd
import statsmodels.stats.diagnostic
```

*IBM* Stock Data from yahoo finance

```python
startDate = datetime.datetime(2022, 10, 1)
endDate = datetime.datetime(2022, 12, 15)
data = yahooFinance.Ticker("IBM")
```

```python
ibm_data = data.history(start=startDate, end=endDate)
ibm_data
```

```python
series_closing = ibm_data["Close"]
```

```python
series_closing.plot(title="Closing Price of IBM stocks")
```

Analysis of Co-relation in the IBM Closing price time series

```python
pd.plotting.autocorrelation_plot(series_closing)
```

LjungBox Test on the Closing Series

```python
dt= statsmodels.stats.diagnostic.acorr_ljungbox(series_closing, lags=None, boxpierce=False, model_df=0, period=None, return_df=True, auto_lag=False)
dt
```

```python
from statsmodels.tsa.stattools import adfuller
```

Adfuller Test on the Closing Series

```python
ad_fullter_test_data = adfuller(series_closing)
ad_fullter_test_data
```

p_value > .05 so it is non stationary

Differencing the time series

```python
ibm_data['diff'] = ibm_data["Close"].diff()
```

Performing Adfuller Test on the difference

```python
ad_fullter_test_data = adfuller(ibm_data['diff'].dropna())
ad_fullter_test_data
```

p_val is less that 0.05. series is stationary

```python
data = pd.read_csv('/content/drive/MyDrive/TSA_BDA_2024/Lab4/Retail_sales_of_electricity_United_States_monthly.csv', skiprows=4)
```

```python
data.dropna(inplace=True)
data['DatePart'] = pd.to_datetime(data['Month'], format='%b %Y', errors='coerce')
data.rename(columns = {"DatePart" : "Date"}, inplace=True)
data.set_index("Date", inplace=True)
data.drop(columns=["Month"], inplace=True)
```

```python
data.plot()
```

Autocorelation plot of the "All sectors" time series

```python
pd.plotting.autocorrelation_plot(data['all sectors million kilowatthours'])
```

Differencing the time series with a lag of 12

```python
data['diff'] = data["all sectors million kilowatthours"].diff(12)
```

Adfuller test on the differences time series

```python
ad_fullter_test_data = adfuller(data['diff'].dropna())
ad_fullter_test_data
```

p_val < 0.05 . series is stationary

Plot of the differenced time series :

```python
data['diff'].plot()
```

LjungBox test on the differenced Time Series - all sectors data

```python
dt= statsmodels.stats.diagnostic.acorr_ljungbox(data['diff'].dropna(), lags=None, boxpierce=False, model_df=0,
                                                period=None, return_df=True, auto_lag=False)
dt
```
